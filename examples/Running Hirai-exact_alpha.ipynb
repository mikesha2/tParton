```python
from tparton.t_evolution import evolve
import matplotlib.pyplot as plt
```

```python
import numpy as np
from scipy.special import gamma
# From Gehrmann (https://arxiv.org/pdf/hep-ph/9512406)
n = 3000
x = np.power(10, np.linspace(np.log10(1/3000), 0, n))
x = np.concatenate(([0], x))

def A(a, b, g, rho):
    return (1 + g * a / (a + b + 1)) \
        * gamma(a) * gamma(b+1) / gamma(a + b + 1) \
        + rho * gamma(a + 0.5) * gamma(b + 1) / gamma(a + b + 1.5)

def pdf_u(x, eta_u=0.918, a_u=0.512, b_u=3.96, gamma_u=11.65, rho_u=-4.60):
    return eta_u / A(a_u, b_u, gamma_u, rho_u) * np.power(x, a_u) * np.power(1-x, b_u) * (1 + gamma_u * x + rho_u * np.sqrt(x))

def pdf_d(x, eta_d=-0.339, a_d=0.780, b_d=4.96, gamma_d=7.81, rho_d=-3.48):
    return pdf_u(x, eta_d, a_d, b_d, gamma_d, rho_d)

u = np.stack((x, pdf_u(x))).T
d = np.stack((x, pdf_d(x))).T
u_evolved = evolve(u, Q0_2=4, Q2=200, l_QCD=0.231, n_f=4, n_t=100, morp='minus', logScale=True, alpha_num=True)
d_evolved = evolve(d, Q0_2=4, Q2=200, l_QCD=0.231, n_f=4, n_t=100, morp='minus', logScale=True, alpha_num=True)
```

```python
fig, ax = plt.subplots()
ax.plot(x, u[:,1]+d[:,1])
ax.plot(x, u_evolved[1]+d_evolved[1])
ax.set_xscale('log')
plt.axis([1e-3,1,0,0.21])
```

```python
x = u[:,0]
y = u[:,1]-d[:,1]
diff = np.stack((x,y)).T
output_plus = evolve(diff, Q0_2=4, Q2=200, l_QCD=0.231, n_f=4, n_t=500, morp='plus', logScale=True, verbose=True, alpha_num=True)
output_minus = evolve(diff, Q0_2=4, Q2=200, l_QCD=0.231, n_f=4, n_t=500, morp='minus', logScale=True, verbose=True, alpha_num=True)
```

```python
x = output_plus[0]
output = (output_plus[1] - output_minus[1]) / 2
plt.plot(x, output)
plt.xscale('log')
plt.xlim([2/3000, 1])
plt.ylim([0, 0.0002])
```

```python
np.savez('hirai-exact_alpha.npz', x=x, upd=u[:,1]+d[:,1], u_evolved=u_evolved[1], d_evolved=d_evolved[1], output_plus=output_plus[1], output_minus=output_minus[1], output=output)
```
